```python
# Importing Libraries
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r'C:\Users\ashik\OneDrive\Desktop\Saba Project\netflix1.csv')
data
```

```python
# Checking for null values
for col in data.columns:
    data_missing = np.mean(data[col].isnull())
    print('{}-{}%'.format(col, round(data_missing*100)))
```

```python
# Clean show_id column
data['show_id'] = data['show_id'].str.replace('s','').astype(int)
data
```

```python
data.info()
```

```python
# Change date_added Column datatype
data['date_added'] = pd.to_datetime(data['date_added'])
data.info()
```

```python
#adding three new columns with day ,month, year
data['year_added'] = data['date_added'].dt.year
data['month_added'] = data['date_added'].dt.month_name()
data['day_added'] = data['date_added'].dt.day_name()
data
```
